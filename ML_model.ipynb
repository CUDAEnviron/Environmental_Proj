### Begin to explore correlation
We added columns to show 'gdp per capita' based on each country's population and gdp numbers and 'co2 per capita' to show how much CO2 is created for person in the country.  Then we can assess if there is a correlation between the gdp of a country and the co2 output.

```python
# create new column for 'gdp per capita'
co2_temp['GDP_per_Capita'] = co2_temp['GDP'] / co2_temp['Population']
co2_temp.head()
```

```python
# create new column for 'co2 per capita'
co2_temp['CO2_per_Capita'] = (co2_temp['CO2'] / co2_temp['Population']) * 10000000
co2_temp.head()
```

To determine a baseline for whether a country is a disproportionate polluter, we broke out the data for 2016, the most recent year in the dataset. We exported it as a csv to get a full look at the data and pick a baseline for "bad" ratio of gdp to co2 per capita.

```python
data_2016 = co2_temp.loc[co2_temp['Year'] == 2016]
data_2016
```

```python
#export new combined dataframe as csv
#data_2016.to_csv("Data2016.csv", index=False)
```

After looking at the 2016 data in Excel, a new column is created to indicate co2 production per unit of gdp, by dividing 'co2 per capita' by 'gdp per capita' (multiplied by 100000 to make the result more readable). A baseline of 300 is chosen, as a cursory analysis shows that numbers between the max and 300 encapsulate countries with economies known to pollute, as well as some countries that are a surprise, so it seems like a good place to begin our analysis. So, countries with a ratio of co2 to gdp over 300, will be labeled as 'high' and those below 300 will be labeled as 'low.

```python
# create new column for 'co2 per unit gdp'
co2_temp['CO2_per_Unit_GDP'] = (co2_temp['CO2'] / co2_temp['GDP_per_Capita']) * 100000
co2_temp
```

```python
# create new columns labelling data as 'high' or 'low' based on co2_per_unit_gdp

co2_temp['Emission_Ratio'] = co2_temp['CO2_per_Unit_GDP'].apply(lambda x: 'High' if x > 300 else 'Low')
co2_temp
```

## Machine Learning Model
### Preprocessing
In this section I worked to encode the categorical data so that it works for the model.  I used the Label Encoder method from the scikit learn library to encode any features where the data was a string, the 'country' and 'emission_ratio' columns.  The data was changed from a string to a number for each unique entry.

```python
# Drop ISO_Code since it's not terribly relevant
co2_temp = co2_temp.drop(['ISO_Code'], axis=1)
```

```python
# encoding categorical data

from sklearn.preprocessing import LabelEncoder

#create labelencoder instance
le = LabelEncoder()

#encode data
co2_temp['Country'] = le.fit_transform(co2_temp['Country'])

co2_temp
```

```python
# encode target feature
from sklearn.preprocessing import LabelEncoder

#create labelencoder instance
le = LabelEncoder()

#encode data
co2_temp['Emission_Ratio'] = le.fit_transform(co2_temp['Emission_Ratio'])
co2_temp
```

### Feature Selection
The X variable, or independent variable, holds all the features that effect the y variable, or dependent variable,  'emission_ratio.'  The X features include 'Country', 'Year', 'CO2', 'Population', 'GDP', 'Temperature', 'GDP_per_Capita', 'CO2_per_Capita', and 'CO2_per_unit_GDP'.  The Emission_Ratio indicates the relationship for each country between GDP output and CO2 emissions per person.  The Emission_Ratio is effected for each country by each feature held in the X variable.

```python
# define the feature set
X = co2_temp.drop('Emission_Ratio', axis=1)
y = co2_temp['Emission_Ratio']
```

```python
y.value_counts()
```

### Split data into training and testing
The data was split into training and testing sets using the train_test_split method from the scikit learn library.

```python
# the dataset is split into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)

```

### Model choice and testing
We chose a balanced random forest model classifier for this analysis.  A BRM was chosen because random forest models are efficient, are not prone to overfitting, and are able to rank feature importance. Howver, a BRM is something of a black box algorithm, so there is little control over what the model does.

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=1)

# Fitting the model
brf.fit(X_train, y_train)
```

### Validate the model
Using the test data to validate the model shows that this model performs with 99% accuracy.  The classification report, furthermore, shows that there is high accuracy for both the high and low classes, for a well balanced model.

```python
# make predictions

y_pred = brf.predict(X_test)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
